Resolva, utilizando PINNs, a EDO a seguir. Faça uso dos dados de treinamento abaixo também, minimizando a perda total formada pelo erro médio quadrado deles e da EDO. Depois compare com a solução analítica:

\begin{equation*}
    \frac{dy}{dt} = y-t^2-1 \hspace{10pt} 0\le t \le 3, \hspace{10pt} y(0)=0.5
\end{equation*}

Solução analítica:

\begin{equation*}
        y(t) = (t+1)^2+2-\frac{5e^t}{2}
\end{equation*}

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Dense,Input
```

```python
def solucao_exata(t):
    return (t+1)**2 + 2 -5/2*np.exp(t)

entradas_treinamento = tf.convert_to_tensor(np.array((0.4,1.3,1.7,2.5)).reshape(-1,1),dtype=tf.float32)
saidas_treinamento = solucao_exata(entradas_treinamento)
x_fisico = tf.linspace(0.,3.,300)
plt.plot(entradas_treinamento[:,0],saidas_treinamento[:,0],"bo",label="Pontos de Treinamento")
plt.plot(np.linspace(0,3),solucao_exata(np.linspace(0,3)),label="Curva real")
plt.legend()
plt.show()
```
